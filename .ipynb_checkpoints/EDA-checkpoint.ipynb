### Data Preparation
* Dropping irrelevant columns and rows
* Identifying duplicated columns
* Renaming columns
* Feature creations

```python
#Example of dropping columns
#df.dropna(['Opening date'], axis = 1)
```

```python
df_1 = df[['coaster_name', 
            #'Length', 'Speed', 
            'Location', 'Status', 
       #'Opening date','Type', 
        'Manufacturer', 
            #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
      # 'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
            #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
            #'height_value', 'height_unit', 
            'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df_1.shape
```

```python
df_1.dtypes
```

```python
#Converting the opening_date_clean to date type
df_1["opening_date_clean"] = pd.to_datetime(df_1['opening_date_clean'])
```

```python
df_1.rename(columns = {'coaster_name':'Coaster_Name',
                    'year_introduced':'Year_Introduced',
                    'latitude':'Latitude',
                    'longitude':'Longitude',
                    'opening_date_clean':'Opening_date_clean',
                    'speed_mph':'Speed_mph',
                    'height_ft':'Height_ft'}, inplace = True)
```

```python
#Checking for missing values
df_1.isna().sum()
```

```python
#Checking for duplicates
df_1.duplicated() #Ignores the first row which is duplicated
```

```python
#Perform df_1.loc[df_1.duplicated()] to see what is duplicated
df_1.loc[df_1.duplicated()]
```

```python
#Checking for duplicates at a subset level
df_1.loc[df_1.duplicated(subset = ['Coaster_Name'])]
```

```python
#Checking example of duplicated subset
df_1.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
#Removing duplicates with the number certain of columns that are the same
#df_1.loc[df_1.duplicated(subset = ['Coaster_Name', 'Location', 'Opening_date_clean'])]
#Locate where values are not duplicates by adding ~ and save it off as df_1
df_1 = df_1.loc[~df_1.duplicated(subset = ['Coaster_Name', 'Location', 'Opening_date_clean'])].reset_index(drop = True)
df_1
```

```python
df_1.shape
```

### Feature Relationships
  * Scatter plot
  * Heatmap correlation
  * Pairplot
  * Groupby comparisons

```python
#Scatter plot using matplotlib
df_1.plot(
    kind = 'scatter',
    x = 'Speed_mph',
    y = 'Height_ft',
    title = 'Coaster Speed vs Height'
)
plt.show()
```

```python
#Scatter plot using seaborn
ax = sns.scatterplot(
    x = 'Speed_mph',
    y = 'Height_ft',
    hue = 'Year_Introduced',
   data = df_1
)
ax.set_title('Coaster Speed vs Height')
plt.show()
```

```python
#Pairplot
sns.pairplot(df_1, vars = ['Year_Introduced', 'Speed_mph', 
                           'Height_ft', 'Inversions_clean', 'Gforce_clean'],
            hue = 'Type_Main')
plt.show()
```

```python
#Comparing Features Using Correlation
df_corr = df_1[['Year_Introduced', 'Speed_mph', 
    'Height_ft', 'Inversions_clean', 'Gforce_clean']].dropna().corr()
df_corr
```

```python
#Heatmap
sns.heatmap(df_corr, annot= True)
```

### Step 5 : Ask a Question About the Data
* Try to answer a question you have about the data using using a plot or a statistic \
What are the locations with the fastest roller costers(minimum of 10)?


```python
df_1.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values('mean')['mean']
```

```python
ax = df_1.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values('mean')['mean'] \
  .plot(kind = 'barh', figsize = (12,5), title = 'Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
